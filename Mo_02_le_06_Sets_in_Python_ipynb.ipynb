##Methods that modify sets

```python
s = {1,2,3,4}
s.add(5)
print(s)
```

```python
s = {1,2,3,4}
s.update({7,8,9})
print(s)
```

```python
s = {1,2,3,4}
s.remove(1)
print(s)
# s.remove(5) erroe
s.discard(5)
print(s)
```

```python
s = {1,2,3,4}
print(s.pop())
print(s)
```

```python
s = {1,2,3,4}
s.clear()
print(s)
```

##set algebra method

```python
s = {1,2,3}
t = {3,4,5}
print(s.union(t))
print(s.intersection(t))
print(s.difference(t))
print(s.symmetric_difference(t))
```

```python
s = {1,2,3}
t = {3,4,5}
s.update(t)
print(s)
```

```python
s = {1,2,3}
t = {3,4,5}
s.intersection_update(t)
print(s)
```

```python
s = {1,2,3}
t = {3,4,5}
s.difference_update(t)
print(s)
```

```python
s = {1,2,3}
t = {3,4,5}
s.symmetric_difference_update(t)
print(s)
```

```python
s = {1,2,3}
t = {1,2}
print(t.issubset(s))
print(s.issuperset(t))
print(s.isdisjoint(t))
```

```python
print(hash(1))
print(hash(1.0))
print(hash(1+0j))
print(hash(True))
```

```python
{1,1.0,True,1 + 0j}
```

```python
print(hash(0))
print(hash(0.0))
print(hash(0+0j))
print(hash(False))
```

```python
print(hash(tuple()))
print(hash(None))
```
