###Dictionary Copying

```python
d1 = {'a':1,'b':[2,3]}
d2 = d1 #just reference
print(id(d1),id(d2))
d2['a']=200
d2['b'].append(500)
print(d1)
print(d2)
```

```python
d1 = {'a':1,'b':[2,3]}
d2 = d1.copy()   #shallow copy
print(id(d1),id(d2))
d2['a']=200
d2['b'].append(500)
print(d1)
print(d2)
print(id(d1['b']), id(d2['b'])) # it will be same
```

```python
import copy
d1 = {'a':1,'b':[2,3]}
d2 = copy.deepcopy(d1)   #deep copy
print(id(d1),id(d2))
d2['a']=200
d2['b'].append(500)
print(d1)
print(d2)
print(id(d1['b']), id(d2['b'])) # it will be same
```

##defaultdict

```python
d = {'a':1,'b':2}
print(d['a'])
print(d['b'])
print(d['c']) # here it will be error as it is not present.....but to deal with this error we use defalutdict.
#...............It's very imp. in graph in comptition programming as we don't want to populate the dictionary
# with all node of graph. Obviously we will want to update it by itself with it's default value when needed and
# later we will change according to our requirements.
```

```python
from collections import defaultdict

d = defaultdict(int)
print(d)
d['a']=1
d['b']=2
print(d)
print(dict(d))
print(d['a'])
print(d['b'])
print(d['c'])
print(dict(d)) # here by default key 'c' with value 0 will be updated.
```

###combine two dictionary if a key appears in both sum the value.

```python
d1  = {'a': 1, 'b': 2, 'c': 3}
d2  = {'a': 4, 'b': 7, 'd': 6}
ans = d1.copy()
for k, v in d2.items():
  if k in ans:
    ans[k]+=v
  else:
    ans[k]=v
print(ans)
```
